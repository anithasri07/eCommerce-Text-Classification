```python
import spacy

nlp= spacy.load("en_core_web_sm")

def preprocess(text):
    filtered_token= []
    doc= nlp(text)
    for token in doc:
        if token.is_punct or token.is_stop:
            continue
        filtered_token.append(token.lemma_)
    return " ".join(filtered_token)
```

```python
df['processed_text'] = df['text'].apply(preprocess)
```
